```python
import sys
import numpy as np

import openmm
from openmm import app
from openmm import unit
```

```python
field = 'field.xml'
config = 'config.pdb'

forcefield = app.ForceField(field)
pdb = app.PDBFile(config)
```

```python
system = forcefield.createSystem(pdb.topology)
```

```python
for i, f in enumerate(system.getForces()):
    name = f.getName()
    if ('BondForce' in name):
        print(f'{name:<20} {f.getNumBonds():4d} bonds')
    elif ('AngleForce' in name):
        print(f'{name:<20} {f.getNumAngles():4d} angles')
    elif ('TorsionForce' in name):
        print(f'{name:<20} {f.getNumTorsions():4d} torsions')
    else:
        print(name)
```

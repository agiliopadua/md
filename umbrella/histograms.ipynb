```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [7, 5]
```

```python
r0min = 2.5
r0max = 10.0
dist = np.arange(r0min, r0max + 0.1, 0.5)
```

```python
rmin = 2.5
rmax = 12.5
nbin = 200
d = np.linspace(rmin, rmax, nbin)
```

```python
fig, ax = plt.subplots()

for r0 in dist:
    ts = np.loadtxt(f'12.8ns/tseries_{r0}.dat', usecols=1)
    h = np.histogram(ts, bins = nbin, range=[rmin, rmax])
    ax.plot(d, h[0])

plt.xlabel("r / A")
plt.show()
```

```python
fig, ax = plt.subplots()

r, pmf = np.loadtxt('12.8ns/wham.out', usecols=(0,1), unpack=True)
ax.plot(r, 4.184*(pmf - 4), linewidth=2)

plt.xlim([2.5, 10.0])

plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels

plt.xlabel("$r$ / $\AA$")
plt.ylabel("PMF / kJ mol$^{-1}$")
plt.savefig('pmf.png')
plt.show()
```
